## 1. Histogram Equalization
Histogram is the method to enhance image contrast.
Please show the image grayscale histogram and do the equlization, and show the image result and histogram equalization result.

```python
img_ori = cv2.imread("input\lena.bmp", cv2.IMREAD_GRAYSCALE) # Read image as gray.

img_equa = cv2.equalizeHist(img_ori)

fig, axes = plt.subplots(2, 2, figsize=(12, 6))
axes[0,0].imshow(img_ori, cmap='gray')
axes[0,0].axis('off')
axes[0,1].hist(img_ori.ravel(), bins=256, range=(0, 256), color='blue')
axes[1,0].imshow(img_equa, cmap='gray')
axes[1,0].axis('off')
axes[1,1].hist(img_equa.ravel(), bins=256, range=(0, 256), color='black')

plt.tight_layout()
plt.show()

# save result
cv2.imwrite('output/img_ori.bmp', img_ori)
cv2.imwrite('output/img_equa.bmp', img_equa)
```

## 2.Mean blur, Gaussian blur, Median blur
Blur is a method to deal with the noise on the image. Please input the image and use different kernel size to show blur result.

```python
img_noise = cv2.imread("input\lena_noise.png", cv2.IMREAD_GRAYSCALE)

# Blur the image by three methods.
img_mean_blur = cv2.blur(img_noise, (5, 5))
img_gaussian_blur = cv2.GaussianBlur(img_noise, (5, 5), 1)
img_median_blur = cv2.medianBlur(img_noise, 5)

fig, axes = plt.subplots(1, 3, figsize=(12, 6))
axes[0].imshow(img_mean_blur, cmap='gray')
axes[0].axis('off')
axes[0].set_title('Mean_blur')
axes[1].imshow(img_gaussian_blur, cmap='gray')
axes[1].axis('off')
axes[1].set_title('Gaussian_blur')
axes[2].imshow(img_median_blur, cmap='gray')
axes[2].axis('off')
axes[2].set_title('median_blur')

plt.show()

# Save the results
cv2.imwrite('output/img_mean_blur.bmp', img_mean_blur)
cv2.imwrite('output/img_gaussian_blur.bmp', img_gaussian_blur)
cv2.imwrite('output/img_median_blur.bmp', img_median_blur)
```

## 3. Binary
Binary is a method to transfer greyscale to 0 (black) and 255 (white) only.  
Please do the image binary to seperate pad and background.  


```python
binary1 = cv2.imread('input/binary_task1.bmp', cv2.IMREAD_GRAYSCALE)
binary2 = cv2.imread('input/binary_task2.bmp', cv2.IMREAD_GRAYSCALE)
binary3 = cv2.imread('input/binary_task3.bmp', cv2.IMREAD_GRAYSCALE)

# Do the image binary to seperate pad and background
_, thres1 = cv2.threshold(binary1, 127, 255, cv2.THRESH_BINARY)
_, thres2 = cv2.threshold(binary2, 150, 255, cv2.THRESH_BINARY)
_, thres3 = cv2.threshold(binary3, 100, 255, cv2.THRESH_BINARY)

fig, axes = plt.subplots(2, 3, figsize=(12, 6))
axes[0,0].imshow(binary1, cmap='gray')
axes[0,0].axis('off')
axes[1,0].imshow(thres1, cmap='gray')
axes[1,0].axis('off')
axes[0,1].imshow(binary2, cmap='gray')
axes[0,1].axis('off')
axes[1,1].imshow(thres2, cmap='gray')
axes[1,1].axis('off')
axes[0,2].imshow(binary3, cmap='gray')
axes[0,2].axis('off')
axes[1,2].imshow(thres3, cmap='gray')
axes[1,2].axis('off')

# save result
cv2.imwrite('output/binary_task1.bmp', thres1)
cv2.imwrite('output/binary_task2.bmp', thres2)
cv2.imwrite('output/binary_task3.bmp', thres3)
```
